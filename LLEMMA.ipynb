# **Step 1: Load LLEMMA model in 4-bit form**

```python
# https://medium.com/@rakeshrajpurohit/model-quantization-with-hugging-face-transformers-and-bitsandbytes-integration-b4c9983e8996
path = "EleutherAI/llemma_7b"
tokenizer = AutoTokenizer.from_pretrained(path)

compute_dtype = getattr(torch, "float16")
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=False,
)
device_map = {"": 0}

model = AutoModelForCausalLM.from_pretrained(
    path,
    device_map=device_map,
    torch_dtype=torch.bfloat16,
    quantization_config=bnb_config
)
tokenizer.padding_side = 'right'
```

# **Step 2: Load & Preprocess Dataset**
  1. Use the first cell to import the raw Math SAT dataset
  2. Use the second cell to import the filtered Math SAT dataset
  3. Use the third cell to import the math competition dataset for fine-tuning

**Note:** All cells should not be loaded, as that would just default to always using the math competition dataset

```python
# https://huggingface.co/datasets/ndavidson/sat-math-chain-of-thought
ds = load_dataset("ndavidson/sat-math-chain-of-thought")["train"]
system_msg = "Answer the following math problem, providing the correct answer and corresponding letter."

def create_conversation(sample):
  return {
    "messages": [
      {"role": "system", "content": system_msg},
      {"role": "user", "content": sample["question"]},
      {"role": "assistant", "content": sample["answer"]}
    ]
  }

ds = ds.filter(lambda example: example["is_correct"] == True)
print(f"There are {len(ds)} correct questions and answers")

ds = ds.map(create_conversation, remove_columns=ds.features,batched=False)

dataset_split = ds.train_test_split(test_size=0.2) # 80% train, 20% test + validation

train_dataset = dataset_split['train']
test_dataset = dataset_split['test']

train_ds = DatasetDict({
    'train': train_dataset,
    'test': test_dataset
})
```

```python
# https://huggingface.co/datasets/knazarali3/group2_processed_sat_math_cot
ds = load_dataset("knazarali3/group2_processed_sat_math_cot")
system_msg = "Answer the following math problem, providing the correct answer and corresponding letter."

def create_conversation(sample):
  return {
    "messages": [
      {"role": "system", "content": system_msg},
      {"role": "user", "content": sample["question"]},
      {"role": "assistant", "content": sample["answer"]}
    ],
    "id": sample["id"]
  }

for i in ds.keys():
  ds[i] = ds[i].map(create_conversation, remove_columns=ds[i].features,batched=False)

train_dataset = ds['train']
test_dataset = ds['test']

train_ds = DatasetDict({
    'train': train_dataset,
    'test': test_dataset
})
```

```python
# https://huggingface.co/datasets/jeggers/competition_math?row=61
ds = load_dataset("jeggers/competition_math", "original")
system_msg = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

    ### Instruction:
    The input contains a math problem. Carefully solve the problem step-by-step with only logical/mathematical steps.

    Please provide the answer in the following format: Begin the response with 'Answer = //boxed' with the final answer in the box. Do not include anything else in the final response.
    The final answer should be an integer or decimal number.

    ### Input:
    {}

    ### Response:

    """

def create_conversation(sample):
  return {
    "messages": [
      {"role": "system", "content": system_msg.format(sample["problem"])},
      {"role": "user", "content": sample["problem"]},
      {"role": "assistant", "content": sample["solution"]}
    ],
    "level": sample["level"],
    "type": sample["type"]
  }

ds['train'] = ds['train'].filter(lambda example: example["level"] != "Level ?")
print(f"There are {len(ds['train'])+len(ds['test'])} addition questions and answers")

for i in ds.keys():
  ds[i] = ds[i].map(create_conversation, remove_columns=ds[i].features,batched=False)

train_ds = DatasetDict({
    'train': ds['train'],
    'test': ds['test']
})
```

# **Step 3: Setup Model for Training**

```python
model, tokenizer = setup_chat_format(model, tokenizer)
```

```python
peft_config = LoraConfig(
        lora_alpha=32,
        lora_dropout=0.05,
        r=32,
        bias="none",
        target_modules=["q_proj", "k_proj", "v_proj", "dense"],
        task_type="CAUSAL_LM",
)
```

```python
output_dir = f'./outputs'

args = TrainingArguments(
    output_dir=output_dir,              # output directory
    num_train_epochs=1,                 # number of epochs to train
    per_device_train_batch_size=1,      # Per device batch size to be loaded in device
    gradient_accumulation_steps=4,      # Gradient accumulation steps for mini-batches
    gradient_checkpointing=True,        # Gradient checkpoint
    optim="adamw_torch_fused",
    logging_steps=25,                   # Logging steps
    save_strategy="steps",              # Save strategy to be steps, can also be epoch
    learning_rate=2e-4,
    fp16=True,                          # fp16 to be loaded and if your gpu supports bf16 then use that
    max_grad_norm=0.3,
    warmup_ratio=0.03,
    lr_scheduler_type="constant",
    max_steps=4000,                     # Max steps will override the training length
    save_steps=100,                     # Save checkpoint after every save_steps
    overwrite_output_dir = 'True'      # will override the dir content
)
```

# **Step 4: Train Model**

```python
max_seq_length = 1024 # max sequence length for model and packing of the dataset
trainer = SFTTrainer(
    model=model,
    args=args,
    train_dataset=train_ds['train'],
    peft_config=peft_config,
    max_seq_length=max_seq_length,
    tokenizer=tokenizer,
    packing=True,
    dataset_kwargs={
        "add_special_tokens": False,  # We template with special tokens
        "append_concat_token": False, # No need to add additional separator token
    }
)

# start training, the model will be automatically saved to the hub and the output directory, resume_from_checkpoint=True
trainer.train(resume_from_checkpoint = True)

# save model
trainer.save_model()
```

```python
del ds
del train_dataset
del test_dataset
```
